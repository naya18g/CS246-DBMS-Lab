```python
import glob
import pandas as pd
num = 27
bt1 = []
files = glob.glob("C:/database-07-mar-2022/database-07-mar-2022/course-wise-students-list/bt/*.csv")
bt2 = [pd.read_csv(f, header=None, sep=",") for f in files]
bt1 = [f[75:-4] for f in files]

for j in range(num):

    n = bt2[j].shape[0]
    a = [bt1[j] for i in range(n)]

    bt2[j].insert(1, "cid", a, True)

bt = pd.concat(bt2,ignore_index=True)
bt.rename(columns = {0:'serial_number',1:'roll_number',2:'name',3:'email'}, inplace = True)
bt.to_csv('BT.csv', header=False, index=False)
```
